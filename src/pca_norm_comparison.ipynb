```python
#First, we'll do a PCA reduction of our first participant without norming the data

#round the array
rw_1_rounded = np.around(rw_1, decimals = 15)


#pca + visualize rw_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
rw_1_components = pca.fit_transform(rw_1_rounded)

#graph 2-d PCA of rw_1


rw_1_x = rw_1_components[:, 0]
rw_1_y = rw_1_components[:, 1]
rw_1_t = rw_1[:,0]
plt.scatter(rw_1_x, rw_1_y, c=rw_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Real World, PCA")
plt.show()
```

```python
#Now, we'll try using a standard pre-processing tool on the same data, max normalization 

#round the array
rw_1_rounded = np.around(normalize(rw_1, axis = 0, norm = 'max'), decimals = 15)


#pca + visualize rw_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
rw_1_components = pca.fit_transform(rw_1_rounded)

#graph 2-d PCA of rw_1
rw_1_x = rw_1_components[:, 0]
rw_1_y = rw_1_components[:, 1]
rw_1_t = rw_1[:,0]
plt.scatter(rw_1_x, rw_1_y, c=rw_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Real World, Normed PCA")
plt.show()
```

```python
#Now, we'll try using a different preprocessing tool - we will scale the data to zero mean and unit variance


#round the array
rw_1_rounded = np.around(scale(rw_1), decimals = 15)
 

#pca + visualize rw_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
rw_1_components = pca.fit_transform(rw_1_rounded)

#graph 2-d PCA of rw_1
rw_1_x = rw_1_components[:, 0]
rw_1_y = rw_1_components[:, 1]
rw_1_t = rw_1[:,0]
plt.scatter(rw_1_x, rw_1_y, c=rw_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Real World, Scaled PCA")
plt.show()
```

```python
#Now, we'll try using a different preprocessing tool - we will scale the data to range [0,1]

#make the scaler
min_max_scaler = MinMaxScaler()

#round the array
rw_1_rounded = np.around(min_max_scaler.fit_transform(rw_1), decimals=15)
 

#pca + visualize rw_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
rw_1_components = pca.fit_transform(rw_1_rounded)

#graph 2-d PCA of rw_1
rw_1_x = rw_1_components[:, 0]
rw_1_y = rw_1_components[:, 1]
rw_1_t = rw_1[:,0]
plt.scatter(rw_1_x, rw_1_y, c=rw_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Real World, [0,1] Scaled PCA")
plt.show()
```

```python
#First, pure PCA of Subject 1's VR:

vr_1_rounded = np.around(vr_1, decimals = 15)


#pca + visualize vr_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
vr_1_components = pca.fit_transform(vr_1_rounded)

#graph 2-d PCA of vr_1
vr_1_x = vr_1_components[:, 0]
vr_1_y = vr_1_components[:, 1]
vr_1_t = vr_1[:,0]
plt.scatter(vr_1_x, vr_1_y, c=vr_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Virtual World, PCA")
plt.show()


```

```python
#Now, let's normalize the VR run

vr_1_rounded = np.around(normalize(vr_1, axis = 0, norm = 'max'), decimals = 15)


#pca + visualize vr_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
vr_1_components = pca.fit_transform(vr_1_rounded)

#graph 2-d PCA of vr_1
vr_1_x = vr_1_components[:, 0]
vr_1_y = vr_1_components[:, 1]
vr_1_t = vr_1[:,0]
plt.scatter(vr_1_x, vr_1_y, c=vr_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Virtual World, Normed PCA")
plt.show()
```

```python
#Now, let's scale the VR run

vr_1_rounded = np.around(scale(vr_1), decimals = 15)


#pca + visualize vr_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
vr_1_components = pca.fit_transform(vr_1_rounded)

#graph 2-d PCA of vr_1
vr_1_x = vr_1_components[:, 0]
vr_1_y = vr_1_components[:, 1]
vr_1_t = vr_1[:,0]
plt.scatter(vr_1_x, vr_1_y, c=vr_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Virtual World, Scaled PCA")
plt.show()
```

```python
#and the ranged scaling: 


#make the scaler
min_max_scaler = MinMaxScaler()

#round the array
vr_1_rounded = np.around(min_max_scaler.fit_transform(vr_1), decimals=15)

#pca + visualize vr_1 in 2 dimensions
pca = decomposition.PCA(n_components = 2)
vr_1_components = pca.fit_transform(vr_1_rounded)

#graph 2-d PCA of vr_1
vr_1_x = vr_1_components[:, 0]
vr_1_y = vr_1_components[:, 1]
vr_1_t = vr_1[:,0]
plt.scatter(vr_1_x, vr_1_y, c=vr_1_t, cmap = "inferno")
plt.colorbar()
plt.title("Subject 1: Virtual World, Scaled PCA")
plt.show()
```

```python
#Now, we'll do the same comparison of norming but this time, we'll use the VR and the Real world data together


#stack our datasets: 

stacked_data = np.vstack((rw_1, vr_1))
data = np.around(stacked_data, decimals = 15)

#Perform the PCA
pca = decomposition.PCA(n_components=2)
stacked_2_components = pca.fit_transform(data)

#Unstack the datasets
rw_1_stacked_2 = stacked_2_components[0:len(rw_1_rounded)]
vr_1_stacked_2 = stacked_2_components[0+len(rw_1_rounded):len(rw_1_rounded)+len(vr_1_rounded)]

#Split the data into x and y values
rw_1_stacked_2_x = rw_1_stacked_2[:, 0]
rw_1_stacked_2_y = rw_1_stacked_2[:, 1]

vr_1_stacked_2_x = vr_1_stacked_2[:, 0]
vr_1_stacked_2_y = vr_1_stacked_2[:, 1]

#Graph it
#Make  the 2d plot for graphing:

fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(rw_1_stacked_2_x, rw_1_stacked_2_y, c='#F8766D', marker = "o", label = 'Subject 1: Real World')
ax.scatter(vr_1_stacked_2_x, vr_1_stacked_2_y, c='#7C8E00', marker = "^", label = 'Subject 1: Virtual World')
plt.title("Subject 1: VR vs Real World, PCA")
plt.legend(loc = 'best'); 

plt.show()
```

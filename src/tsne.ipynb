```python
#imports

# Scikit-learn module for PCA
from sklearn import decomposition

#bhtsne is the Barnes-Hut implementation of t-SNE (Cython).
#Barnes-hut is based on the original Laurens van der maaten paper, and is more memory efficient than sklearn options

from bhtsne import tsne


# Scikit-learn preprocessing module
from sklearn.preprocessing import normalize
from sklearn.preprocessing import scale
from sklearn.preprocessing import MinMaxScaler

# Custom csv processor for importing data from csv
import utility.csv_process as reader

# Numpy for matrix manipulation
import numpy as np

# For 2d graphing
from matplotlib import pyplot as plt

# For 3d graphing
from mpl_toolkits.mplot3d import Axes3D

# For building paths
from os import path
```

```python
#read csvs into np arrays

rw_1 = np.array(reader.cont_read(path.join('data', 'log_gait_1_RW.csv')), dtype = 'float64')
vr_1 = np.array(reader.cont_read(path.join('data', 'log_gait_1_VR.csv')), dtype = 'float64')

rw_2 = np.array(reader.cont_read(path.join('data', 'log_gait_2_RW.csv')), dtype = 'float64')
vr_2 = np.array(reader.cont_read(path.join('data', 'log_gait_2_VR.csv')), dtype = 'float64')

rw_3 = np.array(reader.cont_read(path.join('data', 'log_gait_3_RW.csv')), dtype = 'float64')
vr_3 = np.array(reader.cont_read(path.join('data', 'log_gait_3_VR.csv')), dtype = 'float64')

rw_4 = np.array(reader.cont_read(path.join('data', 'log_gait_4_RW.csv')), dtype = 'float64')
vr_4 = np.array(reader.cont_read(path.join('data', 'log_gait_4_VR.csv')), dtype = 'float64')

rw_5 = np.array(reader.cont_read(path.join('data', 'log_gait_5_RW.csv')), dtype = 'float64')
vr_5 = np.array(reader.cont_read(path.join('data', 'log_gait_5_VR.csv')), dtype = 'float64')

rw_6 = np.array(reader.cont_read(path.join('data', 'log_gait_6_RW.csv')), dtype = 'float64')
vr_6 = np.array(reader.cont_read(path.join('data', 'log_gait_6_VR.csv')), dtype = 'float64')

rw_7 = np.array(reader.cont_read(path.join('data', 'log_gait_7_RW.csv')), dtype = 'float64')
vr_7 = np.array(reader.cont_read(path.join('data', 'log_gait_7_VR.csv')), dtype = 'float64')

rw_8 = np.array(reader.cont_read(path.join('data', 'log_gait_8_RW.csv')), dtype = 'float64')
vr_8 = np.array(reader.cont_read(path.join('data', 'log_gait_8_VR.csv')), dtype = 'float64')

rw_9 = np.array(reader.cont_read(path.join('data', 'log_gait_8_RW.csv')), dtype = 'float64')
vr_9 = np.array(reader.cont_read(path.join('data', 'log_gait_8_VR.csv')), dtype = 'float64')

rw_10 = np.array(reader.cont_read(path.join('data', 'log_gait_10_RW.csv')), dtype = 'float64')
vr_10 = np.array(reader.cont_read(path.join('data', 'log_gait_10_VR.csv')), dtype = 'float64')
```

```python
#Stack all the data and print the shape

stacked_data = np.vstack((rw_1, vr_1, rw_2, vr_2, rw_3, vr_3, rw_4, vr_4, rw_5, vr_5, rw_6, vr_6, rw_7, vr_7, rw_8, vr_8, rw_9, vr_9, rw_10, vr_10))
data = np.around(normalize(stacked_data, axis = 0, norm = 'max'), decimals = 15)



print(data.shape)

```

```python
#Run the projection using the barnes-hut implementation
projection = tsne(data)
print(projection.shape)
```

```python
#Now, we need to graph that. 

#First, a quick helper function:

def get_coordinates(last, lengths, proj):
    
    k = 0
    x = 0 
    y = 0 
    while k < len(lengths):
        if k <= last:
            y+=lengths[k]
        if k < last:
            x+=lengths[k]
        k+=1

    return (proj[x:y])
```

```python
#Next, we'll create a list of lengths, and then get to work

lengths = [len(rw_1), len(vr_1), len(rw_2), len(vr_2), len(rw_3), len(vr_3), len(rw_4), len(vr_4), len(rw_5), len(vr_5), len(rw_6), len(vr_6), len(rw_7), len(vr_7), len(rw_8), len(vr_8), len(rw_9), len(vr_9), len(rw_10), len(vr_10)]

```

```python
rw_1_stacked = get_coordinates(0, lengths, projection)
vr_1_stacked = get_coordinates(1, lengths, projection)

rw_2_stacked = get_coordinates(2, lengths, projection)
vr_2_stacked = get_coordinates(3, lengths, projection)

rw_3_stacked = get_coordinates(4, lengths, projection)
vr_3_stacked = get_coordinates(5, lengths, projection)

rw_4_stacked = get_coordinates(6, lengths, projection)
vr_4_stacked = get_coordinates(7, lengths, projection)

rw_5_stacked = get_coordinates(8, lengths, projection)
vr_5_stacked = get_coordinates(9, lengths, projection)

rw_6_stacked = get_coordinates(10, lengths, projection)
vr_6_stacked = get_coordinates(11, lengths, projection)

rw_7_stacked = get_coordinates(12, lengths, projection)
vr_7_stacked = get_coordinates(13, lengths, projection)

rw_8_stacked = get_coordinates(14, lengths, projection)
vr_8_stacked = get_coordinates(15, lengths, projection)

rw_9_stacked = get_coordinates(16, lengths, projection)
vr_9_stacked = get_coordinates(17, lengths, projection)

rw_10_stacked = get_coordinates(18, lengths, projection)
vr_10_stacked = get_coordinates(19, lengths, projection)



```
